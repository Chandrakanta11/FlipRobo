# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.

```python
page1 = requests.get('https://www.imdb.com/list/ls055592025/')
imdb= BeautifulSoup(page1.content,"html.parser")
```

```python
#scraping names of the movies

names =[]

for i in imdb.find_all('h3',class_="lister-item-header"):
    names.append(i.get_text().strip("\n").replace("\n","").split("(")[0])
```

```python
#scraping year of the movies

year =[]

for i in imdb.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
```

```python
#scraping rating of the movies

ratings =[]

for i in imdb.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace("\n",""))
    
```

```python
#Making DataFrame

df= pd.DataFrame({'Movie_Names':names,'Year':year,'Rating':ratings})
df
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and ratin

```python
page4= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
men = BeautifulSoup(page4.content,'html.parser')
```

```python
teams=[]
for i in men.find_all("span",class_="u-hide-phablet"):
    teams.append(i.get_text())
```

```python
matches=[]
matches.append((men.find("td",class_="rankings-block__banner--matches")).text)
x=1
for i in men.find_all("td",class_="table-body__cell u-center-text"):
    if x%2==1:
        matches.append(i.text)
    x=x+1
```

```python
points=[]
points.append((men.find("td",class_="rankings-block__banner--points")).text)
x-=1

for i in men.find_all("td",class_="table-body__cell u-center-text"):
    if x%2==0:
        points.append(i.text)
    x=x+1
```

```python
ratings=[]

for i in men.find_all("td",class_="rankings-block__banner--rating u-text-right"):
    ratings.append(i.get_text().replace("\n","").replace(" ",""))
    
for i in men.find_all("td",class_="table-body__cell u-text-right rating"):
    ratings.append(i.get_text())
```

```python
#Making DataFrame

dfm= pd.DataFrame({'Teams':teams,'Matches':matches,'Points':points,'Ratings':ratings})
dfm[0:10]
```

```python
df= pd.DataFrame({'Headline':title,'Time':time,'News link':links})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL


```python
page10 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
rest = BeautifulSoup(page10.content,'html.parser')
```

```python
names =[]

for i in rest.find_all("div",class_="restnt-info cursor"):
    names.append(i.get_text())
    
cousine =[]
for i in rest.find_all("div",class_="detail-info"):
    cousine.append(i.get_text().split('|')[1])
    
loc =[]

for i in rest.find_all("div",class_="restnt-loc ellipsis"):
    loc.append(i.get_text())
    
rating = []

for i in rest.find_all("div",class_="restnt-rating rating-4"):
    rating.append(i.get_text())
    
img = []

for i in rest.find_all("img",class_="no-img"):
    img.append(i.get('data-src'))
```

```python
df= pd.DataFrame({'Restaurant Names':names,'Cousine':cousine,'Location':loc,'Ratings':rating,'Images-url':img})
df
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from 
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank 
ii) Publication
iii) h5-index
 iv) h5-media

```python
page11=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
book= BeautifulSoup(page11.content,"html.parser")
```

```python
rank =[]
for i in book.find_all("td",class_="gsc_mvt_p"):
    rank.append(i.get_text())
    
pub =[]
for i in book.find_all("td",class_="gsc_mvt_t"):
    pub.append(i.get_text())

hindex=[]
for i in book.find_all("a",class_="gs_ibl gsc_mp_anchor"):
    hindex.append(i.get_text())
    
hmedia=[]
for i in book.find_all("span",class_="gs_ibl gsc_mp_anchor"):
    hmedia.append(i.get_text())
    
```

```python
dfb = pd.DataFrame({'Rank':rank,'Publication':pub,'h5-index':hindex,'h5-media':hmedia})
dfb
```
