```python
#Creating a list of IMDB's top 100 movies titles

from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/list/ls055592025/')
head= BeautifulSoup(page.content,"html.parser")
```

```python
#scraping names of the movies

names =[]

for i in head.find_all('h3',class_="lister-item-header"):
    names.append(i.get_text().strip("\n").replace("\n","").split("(")[0])
    

names
```

```python
#scraping year of the movies

year =[]

for i in head.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
year
```

```python
#scraping rating of the movies

ratings =[]

for i in head.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace("\n",""))
    
ratings
```

```python
#Making DataFrame

import pandas as pd
df= pd.DataFrame({'Movie_Names':names,'Year':year,'Rating':ratings})
df
```
