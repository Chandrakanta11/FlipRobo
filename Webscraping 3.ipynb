# Q2- In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.


```python
brand=[]
name=[]
price=[]
re=[]
expd=[]
avail=[]
urls=[]
```

```python
start=0
end=2
for page in range(start,end):
    url =driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
    for i in url:
        urls.append(i.get_attribute("href"))
    
    next_b=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_b.click()
    time.sleep(5)
```

```python
for url in urls:
    driver.get(url)
    time.sleep(1)
    
    try:
        brnd=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        brand.append(brnd.text)
    except NoSuchElementException:
        brand.append('--')
        
for url in urls:
    driver.get(url)
    time.sleep(2)
        
    try: 
        Name=driver.find_element(By.XPATH,'//*[@id="title_feature_div"]/div/h1/span')
        name.append(Name.text)
    except NoSuchElementException:
        name.append("-")
        
for url in urls:
    driver.get(url)
    time.sleep(2)
        
    try:
        pri=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(pri.text)
    except NoSuchElementException:
        price.append("-")
        
for url in urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        Exp=driver.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        expd.append(Exp.text)
    except NoSuchElementException:
        expd.append("-")
```

```python
df2=pd.DataFrame({'Brand':brnad,'Name':name,'Price':price,'Expected Delivery':expd})
df2
```

# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) onwww.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver.get("https://www.flipkart.com/")
time.sleep(2)
close_signin=driver.find_element(By.XPATH, '//button[@class="_2KpZ6l _2doB4z"]')
close_signin.click()
driver.maximize_window()


product=driver.find_element(By.CLASS_NAME, "_3704LK") 
brand=input("Enter the smart phone brand name: ")
product.send_keys(brand)


search=driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search.click()
driver.implicitly_wait(3)
```

```python
product_urls=[]
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//div[@class="_13oc-S"]/div/div/a')   
    for i in url:
        product_urls.append(i.get_attribute("href"))      
    driver.implicitly_wait(2)

Brand_Name=[]
Smart_Phone_Name=[]
Colour=[]
RAM=[]
Storage_ROM=[]
Camera=[]
Display_Size=[]
Battery_Capacity=[]
Price=[]

for i in product_urls:
    driver.get(i)
    driver.implicitly_wait(2)
    read_more=driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')
    read_more.click()
    driver.implicitly_wait(1)
    Brand_Name.append(brand)
    
    
    try:
        phone_name=driver.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        Smart_Phone_Name.append(phone_name.text)
    except NoSuchElementException:
        Smart_Phone_Name.append("_")
        
    try:
        phone_color=driver.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        Colour.append(phone_color.text)
    except NoSuchElementException:
        Colour.append("_")
        
    try:
        phone_RAM=driver.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        RAM.append(phone_RAM.text)
    except NoSuchElementException:
        RAM.append("_")
        
    try:
        phone_ROM=driver.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        Storage_ROM.append(phone_ROM.text)
    except NoSuchElementException:
        Storage_ROM.append("_")
    
    try:
        camera_details=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]')
        Camera.append(camera_details.text)
    except NoSuchElementException:
        Camera.append("_")  
    
    try:
        phone_display_size=driver.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        Display_Size.append(phone_display_size.text)
    except NoSuchElementException:
        Display_Size.append("_")
    
    try:
        phone_battery=driver.find_element(By.XPATH,'//li[@class="_21lJbe" and contains(text(),"mAh")]')
        Battery_Capacity.append(phone_battery.text)
    except NoSuchElementException:
        Battery_Capacity.append("_")

    try:
        phone_price=driver.find_element(By.XPATH,'//div[@class="dyC4hf"]/div[1]/div/div')
        Price.append(phone_price.text)
    except NoSuchElementException:
        Price.append("_")
        
driver.close()
```

```python
df4=pd.DataFrame({'Brand Name':Brand_Name,'Smart Phone Name':Smart_Phone_Name,'Colour':Colour,'RAM':RAM,'Storage ROM':Storage_ROM,'Camera':Camera,'Display Size':Display_Size,'Battery Capacity':Battery_Capacity,'Price':Price,'Product URL':product_urls})
df4
```
